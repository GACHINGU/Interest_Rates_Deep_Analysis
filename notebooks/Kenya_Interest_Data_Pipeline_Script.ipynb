```python
import requests
import pandas as pd

# World Bank API endpoint (Kenya = KEN)
url = "https://api.worldbank.org/v2/country/KEN/indicator/FR.INR.RINR"

params = {
    "format": "json",
    "per_page": 1000
}

# Fetch data
response = requests.get(url, params=params)
data = response.json()

# Build DataFrame
df = pd.DataFrame(data[1])[["date", "value"]]

# Rename columns (clean style)
df = df.rename(columns={
    "date": "year",
    "value": "interest_rate"
})

# Clean data
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["interest_rate"] = pd.to_numeric(df["interest_rate"], errors="coerce")

df = df.dropna()
df["year"] = df["year"].astype(int)

# Sort properly
df = df.sort_values("year").reset_index(drop=True)

# Round values
df["interest_rate"] = df["interest_rate"].round(2)

# Output
print(df.head())
print(df.tail())

# Save
df.to_csv("kenya_interest_rates.csv", index=False)
```
